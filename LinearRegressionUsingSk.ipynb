```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/placement.csv')
```

```python
df.head()
```

```python
# we have to draw scatter plot for LR
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package(in Lpa)')
plt.show()
```

```python
#df.iloc: This accesses rows and columns in the DataFrame by their integer positions (index-based selection).
#0:1 is a slice: it includes the column at index 0, but excludes index 1.
#If you want a single column as a Series, use df.iloc[:, 0] (no colon after the comma).
#If you want a single column as a DataFrame, use df.iloc[:, [0]] (double brackets).


X=df.iloc[:,[0]]
y=df.iloc[:,-1]
```

```python
y
```

```python
X
```

```python
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

**What lr.fit(X_train, y_train) Does:??**

1.Finds the Best-Fit Line

2.Minimizes the Loss Function

3.Stores the Coefficients

```python
X_test
```

```python
y_test
```

This is the actual output (package values) for the test data. These values are used to evaluate how well your model performs.
Index,Package

Scikit-learn's predict() method always expects a 2D array, even for a single sample.
If you don't reshape, you'll get an error . **bold text**

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='black')
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```

```python
m=lr.coef_
b=lr.intercept_
```

```python
m*9.5+b
```
